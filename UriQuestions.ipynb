```python
# 1036: Fórmula de Bhaskara

A, B, C = [float(x) for x in input().split()]
if A != 0:
    delta = B**2 - 4*A*C
    if delta >= 0:
        R1 = (-B + (delta**0.5))/(2*A)
        R2 = (-B - (delta**0.5))/(2*A)
        print('R1 =',"%.5f" %R1)
        print('R2 =',"%.5f" %R2)
    else:
        print("Impossivel calcular")
else:
    print("Impossivel calcular")
```

```python
#1045: Tipos de triângulos

a, b, c = [float(x) for x in input().split()]
numeros = [a, b, c]
ordem = sorted(numeros)
A = ordem[2]
B = ordem[1]
C = ordem[0]

if A >= B + C:
    print("NAO FORMA TRIANGULO")
else:
    if A**2 == (B**2 + C**2):
        print("TRIANGULO RETANGULO")
    elif A**2 > (B**2 + C**2):
        print("TRIANGULO OBTUSANGULO")
    elif A**2 < (B**2 + C**2):
        print("TRIANGULO ACUTANGULO")
    if A == B and B == C:
        print("TRIANGULO EQUILATERO")
    elif A == B and B != C or C == B and A != C:
        print("TRIANGULO ISOSCELES")
```

```python
# 1185 Acima da diagonal secundária da Matriz
O = str(input())
coluna = []
matriz = 3
for i in range(matriz):
    linha =[]
    for j in range(matriz):
        aij = float(input())
        linha.append(aij)
    coluna.append(linha)

diagsec = []
elem = matriz - 1
for a in range(matriz):
    line = coluna[a]
    d = line[0:elem]
    diagsec.extend(d)
    elem = elem -1
if O == 'S':
    vetor = diagsec
    soma = sum(vetor)
    print("%.1f" %soma)
elif O == 'M':
    vetor = diagsec
    media = sum(vetor)/len(vetor)
    print("%.1f"%media)
```

```python
#1435 Matriz Quadrada
n = int(input())
while n != 0:
    for i in range(1,n+1):
        for j in range(1,n+1):
            minimo = i
            if j < minimo:
                minimo = j
            if n - i + 1 < minimo:
                minimo = n-i+1
            if n - j + 1 < minimo:
                minimo = n-j+1
            print('%3d' %minimo, end = '')

            if j < n: 
                print(' ', end = '')
            else: print()
                          
    print('\n', end = '')       
    n = int(input())

```

```python
# 1478 Matriz Quadrada II
n = int(input())
while n != 0:
    for i in range(1,n+1):
        minimo = i
        for j in range(1,n+1):
            if i == j:
                minimo = 1
            if j > i:
                minimo = minimo+1
            if i > j and j!=1:
                minimo = minimo-1
                
            print('%3d' %minimo, end = '')

            if j < n: 
                print(' ', end = '')
            else: print()
                          
    print('\n', end = '')       
    n = int(input())
```

```python
# 1534 Matriz 123
n = int(input())
while True:
    try:
        elem = n
        for i in range(1,n+1):
            for j in range(1,n+1):
                if j == elem:   
                    minimo = 2 
                elif i == j:
                    minimo = 1 
                else:
                    minimo = 3
                print(minimo,end='')

                if j < n: 
                    print('', end = '')
                else: print()
            elem = elem -1                       
        n = int(input())
    except EOFError:
        break
```

```python
# 1557 Matriz Quadrada III
n = int(input())
while n!= 0:
    maximo = 2**(2*n -2)
    count = 0
    while (maximo > 0):
        maximo = maximo//10
        count = count + 1
    
    elemi = 1
    for i in range(1,n+1):
        elemj = elemi*1
        for j in range(1,n+1):  
            minimo = elemj
            elemj = elemj*2
            
            print('%*d' %(count, minimo), end = '')
            if j < n: 
                print(' ', end = '')
            else: print()
        elemi = elemi*2   
    print('\n', end = '')      
    n = int(input())
```

```python
#1214 Acima da Média

C = int(input())
for x in range(C):
    alunos = [int(x) for x in input().split()]
    notas = alunos[1:]
    media = sum(notas)/len(notas)
    def maior(x):
        return x > media
    maximos = list(filter(maior, notas))
    acimamedia = len(maximos)
    total = 100*(acimamedia/len(notas))
    print("%.3f%%" %total)

    
```
